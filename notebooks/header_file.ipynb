```python
import boto3
from dotenv import load_dotenv
load_dotenv()

bucket_name = "innodata-etl-dev"

s3_client = boto3.client('s3')
response = s3_client.list_objects_v2(Bucket=bucket_name)

if 'Contents' in response:
    print(f"Objetos no bucket '{bucket_name}':")
    for obj in response['Contents']:
        print(f"- {obj['Key']}")
else:
    print(f"O bucket '{bucket_name}' está vazio ou não foi encontrado.")
```

```python
import pandas as pd

file_key = "payroll/payroll_mapping/PAYROLL_HEADERS.CSV"
response = s3_client.get_object(Bucket=bucket_name, Key=file_key)

raw_data = response['Body'].read()

try:
    df = pd.read_csv(pd.io.common.BytesIO(raw_data), encoding='latin1', sep=';')
    display(df.head())
    print(df.info())
except Exception as e:
    print(f"Erro: {e}")
```
